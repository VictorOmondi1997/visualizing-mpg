`mpg` dataset contains one row per car model and includes information such as the year the car was made, the number of miles per gallon ("M.P.G.") it achieves, the power of its engine (measured in "horsepower"), and its country of origin.

What is the relationship between the power of a car's engine ("horsepower") and its fuel efficiency ("mpg")? And how does this relationship vary by the number of cylinders ("cylinders") the car has? Let's find out.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```
mpg = pd.read_csv('https://assets.datacamp.com/production/repositories/3996/datasets/e0b285b89bdbfbbe8d81123e64727ff150d544e0/mpg.csv')
mpg.head()
```

```
sns.relplot(data=mpg, kind='scatter', x='horsepower', y='mpg', size='cylinders', hue='cylinders')
plt.show()
```

Cars with higher horsepower tend to get a lower number of miles per gallon. They also tend to have a higher number of cylinders.

 looking at the relationship between how fast a car can accelerate ("acceleration") and its fuel efficiency ("mpg"). Do these properties vary by country of origin ("origin")?

 > Note that the "acceleration" variable is the time to accelerate from 0 to 60 miles per hour, in seconds. Higher values indicate slower acceleration.

```
sns.relplot(data=mpg, kind='scatter', x='acceleration', y='mpg', hue='origin', style='origin')
plt.show()
```

Cars from the USA tend to accelerate more quickly and get lower miles per gallon compared to cars from Europe and Japan.

How has the average miles per gallon achieved by these cars changed over time? Let's use line plots to find out!

```
sns.relplot(data=mpg, kind='line', x='model_year', y='mpg')
plt.show()
```

### Interpretation of this line plot

* The average miles per gallon has generally increased over time.
* The 95% confidence interval for average miles per gallon in 1970 is approximately 16 - 19.5 miles per gallon.
* This plot assumes that our data is a random sample of all cars in the US, Europe, and Japan.

# visualizing how the distribution of miles per gallon has changed over time.

```
sns.relplot(data=mpg, kind='line', x='model_year', y='mpg', ci='sd')
plt.show()
```

We've seen that the average miles per gallon for cars has increased over time, but how has the average horsepower for cars changed over time? And does this trend differ by country of origin?

```
sns.relplot(data=mpg, kind='line', x='model_year', y='horsepower', ci=None, style='origin', hue='origin', markers=True, dashes=False)
plt.show()
```

 we can see that this downward trend in horsepower was more pronounced among cars from the USA.

# Relationship between a car's weight and its horsepower

```
g = sns.relplot(x="weight", y="horsepower", data=mpg,kind="scatter")
g.fig.suptitle('Car weight vs. Horsepower', y= 1.03)
plt.show()
```

It looks like a car's weight is positively correlated with its horsepower.

# the average acceleration for cars in each of the three places of origin.

```
sns.catplot(data=mpg,kind='point', x='origin', y='acceleration', join=False, capsize=0.2)
plt.xticks(rotation=90)
plt.show()
```

Since higher values indicate slower acceleration, it looks like cars from Japan and Europe have significantly slower acceleration compares to the USA.
